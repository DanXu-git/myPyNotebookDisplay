```python
%matplotlib inline
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
import re
import sklearn
```

```python
df = pd.read_excel("../input/8月Hiseq.xlsx")
df.rename(columns={'Flowcell编号(Flowcell No.)' : 'FLOWCELL'}, inplace=True)
df.head()
```

```python
FC_yield = df.groupby('FLOWCELL')['总产量(Total yield of data)'].sum()
FC_yield = pd.DataFrame(FC_yield)
FC_yield.reset_index(inplace=True)
```

```python
df.groupby(['执行人', '项目来源(Source of Project)'])['总产量(Total yield of data)'].sum()
```

```python
df2 = pd.read_excel("../input/DP执行时间.xlsx")
df2.rename(columns={'Unnamed: 1':'Type'}, inplace=True)
df2['FLOWCELL'] = [re.sub("^[AB]","FC",df2['FC'][i].split('_')[-1]) for i in df2.index]
df2.head()
```

```python
df_all = pd.merge(FC_yield, df2, how='left', on='FLOWCELL')
df_all = df_all.drop(df_all.loc[df_all.Type.isnull()].index)
df_all.shape
```

```python
#reg = linear_model.LinearRegression()
#reg.fit([df_all['总产量(Total yield of data)'], df_all['Type'], df_all['区域']], df_all['分析时间'])
df_all.to_csv("../input/August_DP.csv", index=False, encoding="utf-8")
```

```python
DT_Aug = pd.read_excel("../input/DT_Aug.xlsx")
DT_Aug['FC'] = [re.sub("^[AB]","FC",DT_Aug['FC'][i].split('_')[-1]) for i in DT_Aug.index]
DT_Aug = DT_Aug.drop(DT_Aug[DT_Aug.Machine == "Hiseq2500_Normal"].index)
DT_Aug.head()
```

```python
exp_yield = DT_Aug.groupby(['Machine', 'RL', 'Type'])['Yield'].quantile(0.95)
exp_yield = pd.DataFrame(exp_yield).reset_index()
exp_yield['Area'] = pd.Series(['武汉' for i in range(len(exp_yield))])
exp_yield.to_csv("../input/stand_data.csv", index=False, encoding="utf-8")
```

```python
f, ax = plt.subplots(figsize=(8,6))
for i in DT_Aug.Type.unique():
    p1 = sns.distplot(DT_Aug[DT_Aug.Type == i].Yield)
#p1.axis(xmax=40)
p1.legend(labels = DT_Aug.Type.unique())
```

```python
DT_Aug = DT_Aug.drop(DT_Aug[DT_Aug.Coms_time > 40].index)
DT_Aug['Coms_time'].describe()
```

```python
DT_Aug.to_csv("../input/DT_Aug.csv", index=False, encoding="utf-8")
```

```python
corrmat = DT_Aug.corr()
plt.subplots(figsize=(8, 8))
sns.set(font_scale=1.25)
hm = sns.heatmap(corrmat, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 20})
```

```python
from sklearn.cluster import KMeans
Coms_time = DT_Aug[DT_Aug.Type == 'PE'].Coms_time.values.reshape(-1, 1)
x_ = range(len(Coms_time))
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(Coms_time)
```

```python
f, ax = plt.subplots(figsize=(8,6))
plt.ylim([0,50])
fig = sns.boxplot("Area", "Coms_time", data=DT_Aug)
```

```python
plt.figure(figsize=(15,15))
plt.scatter(x_, Coms_time, cmap=kmeans.labels_.astype('str'))
for x,y,tt in zip(x_, Coms_time, DT_Aug[DT_Aug.Type == 'PE'].Area):
    plt.text(x, y, tt, fontsize=16)
```

```python
sns.distplot(DT_Aug.Yield, fit=norm)
fig = plt.figure()
res = stats.probplot(DT_Aug.Yield, plot=plt)
```
